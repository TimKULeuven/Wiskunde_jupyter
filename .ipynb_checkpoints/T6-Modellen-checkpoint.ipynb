# T6 - Modelleren

Een wiskundig model tracht een systeem uit de werkelijkheid te vatten in een wiskundige uitdrukking, die dan met behulp van wiskundige technieken en computers bestudeerd kan worden.

## T6.1 - Waarom zijn modellen zo handig?


### T6.1.1 - Inzicht in de werkelijkheid

Het is duidelijk dat de realiteit uitermate complex is. Stel dat je exact wenst te voorspellen hoe snel
een bepaald bos zal groeien. Hiertoe is het nodig de interne structuur van elke boom te kennen en in vergelijkingen uit te drukken, om te weten te komen hoe voedingsstoffen over de verschillende bomen en takken gedistribueerd
worden. De atmosfeer zal gesimuleerd moeten worden, om te voorspellen hoeveel zonlicht en CO$_2$ elke boom bereikt. De groei van bacteriënpopulaties op de wortels van de bomen moeten worden in kaart gebracht, omdat de stikstoffixatie die deze bacteriën veroorzaken een invloed heeft op de groei van de bomen.

Men kan zo eindeloos doorgaan met processen opsommen die een (kleine) invloed hebben op hoe het bos groeit, en die dus, indien we de werkelijkheid precies willen nabootsen, in vergelijkingen moeten worden gevat. Vooreerst is het onbegonnen werk vergelijkingen op te stellen die bovenstaande fenomenen exact beschrijven, en zelfs indien dit mogelijk zou zijn, is er geen computer voorhanden die het bekomen model aan een aanvaardbare snelheid zou kunnen uitrekenen.

Een essentieel onderdeel van het opstellen van modellen bestaat er dan ook uit *assumpties* te maken. Deze aannames trachten de realiteit te vereenvoudigen, zonder de waarheid te veel geweld aan te doen. Ze scheiden de belangrijke fenomenen van de onbelangrijke, en maken het zo mogelijk een wiskundig model te formuleren voor de vereenvoudigde realiteit die
na het maken van assumpties bekomen werd. Eens men dan, na het maken van assumpties, een vrij eenvoudig wiskundig model bekomt, is het belangrijk na te gaan of de voorspellingen die dit model maakt inderdaad overeenstemmen met wat we in realiteit vaststellen. We gaan dan onze *modelvoorspelling* naast *experimentele data* gaan leggen.

Natuurlijk zullen we na het maken van onze vereenvoudigende assumpties nooit exact kunnen voorspellen hoe elke boom in het bos precies groeit, maar het zou wel mogelijk moeten zijn *ongeveer* te weten te komen hoeveel ton hout er per jaar geproduceerd wordt, hoeveel vierkante kilometer bladgroei er is, of hoeveel ton CO$_2$ per jaar door het bos ongeveer in zuurstof wordt omgezet. Bij het modelleren maken we dus steeds een afweging: we proberen de werkelijkheid met behulp van assumpties net voldoende te vereenvoudigen, opdat we een model bekomen dat de werkelijkheid voldoende goed benadert, en terzelfdertijd eenvoudig genoeg is om met de computer te bestuderen.

Hier komen we tot een eerste verdienste van modellen. De cyclus waarin we assumpties maken en vervolgens nagaan of het resulterende model voldoende met de werkelijkheid strookt, staat ons toe objectief na te gaan welke factoren een grote impact hebben op de groei van het bos, en welke factoren verwaarloosd mogen worden. Indien we een factor toevoegen aan het model, maar de resulterende modelvoorspelling benadert de experimentele data niet beter, kunnen we vermoeden dat deze factor geen grote impact heeft op het onderwerp dat we bestuderen.

In de farmaceutische wereld worden we met extreem complexe probleemstellingen geconfronteerd, omdat we het gedrag van miljarden geneesmiddelmolecules in het raderwerk van het menselijk lichaam willen begrijpen. Toch blijkt (zie [H1 - Farmacokinetiek](./H1-Farmacokinetiek.ipynb)) dat we met zeer sterke assumpties, waarin we een mens bijna als een vat water van een bepaald volume beschouwen, toch al heel wat inzicht kunnen krijgen over de afbraak van geneesmiddelen in het menselijk lichaam. Indien we meer specifieke fenomenen meer precies willen beschrijven, bijvoorbeeld de diffusie van een bepaald geneesmiddel door de darmwand, zijn er natuurlijk andere modellen (zie [H7 - Diffusie](./H7-Diffusie.ipynb)) nodig.


### T6.1.2 - Voorspellingen maken

Eens we een voldoende eenvoudig model hebben, dat voldoende dicht bij de experimentele data, en dus de realiteit, aansluit, kunnen we het aanwenden om te gissen wat de toekomst in bepaalde situaties brengt. Modellen spelen dan ook een essentiële rol bij het in kaart brengen van de klimaatproblematiek, en worden ook gebruikt om de slaagkansen van studenten die aan een opleiding beginnen te berekenen, of om de banen van aanstormende kometen in kaart te brengen. De wiskundige analyse van het model kan bovendien eigenschappen van de realiteit aan het licht brengen die we experimenteel nog niet hebben vastgesteld, omdat we de realiteit nog niet onderworpen hadden aan de condities waarvoor het model erg bijzondere gevolgen voorspeld.

In farmaceutische context gebruikt men computermodellen om te voorspellen welke moleculen potentieel een bepaalde aandoening kunnen bestrijden. Men tracht dan de affiniteit van de kandidaatmolecule voor het biologische target te voorspellen, en zodoende een lang en kostelijk proces van labotesten in te korten. Men kan ook modellen gebruiken om de verspreiding van besmettelijke ziektes in kaart te brengen (zie [H3-Populatiemodellen](./H3-Populatiemodellen.ipynb)), en te voorspellen hoe de verspreiding zich zal verderzetten, en welke gevolgen dat met zich meebrengt.


### T6.1.3 - Goedkoop experimenteren

Een wiskundig model laat ons toe virtueel te experimenteren. We kunnen met behulp van de computer bestuderen welke voorspellingen we krijgen indien we de waarden van enkele randvoorwaarden of parameters in ons model aanpassen. Dit heeft onder meer een grote impact in de (farmaceutische) industrie. Stel bijvoorbeeld dat we, zoals in [H6 - Het Industrieel Productieproces](./H6-HetIndustrieelProductieproces.ipynb), een model hebben dat de productie van een pharmaceutisch relevante molecule in een chemische reactor beschrijft. Als bedrijf wensen we op goedkope, snelle manier zo veel mogelijk van deze molecule aanmaken. Hiertoe moeten we onder meer de temperatuur in de tank, en de snelheid waarmee reactanten aan de tank worden toegevoegd, zo instellen dat er zo veel mogelijk reactieproduct gevormd wordt. Men kan zich inbeelden dat het een erg duur en tijdrovende taak is om een hele hoop experimenten op te starten waar we bij verschillende temperaturen en instroomsnelheden uitmeten hoeveel product er gevormd wordt. Dankzij het model kunnen we in de computer al deze experimenten virtueel laten doorgaan, wat veel sneller en goedkoper is. Wanneer we dan een optimale configuratie hebben gevonden, kunnen we deze in realiteit uitvoeren, en nagaan of de opbrengst inderdaad zo goed is als de computer voorspeld had.

Ziekteverspreidingsmodellen (zoals in [H3-Populatiemodellen](./H3-Populatiemodellen.ipynb)) modelleren de verspreiding van een besmettelijke ziekte over een populatie. Als we willen onderzoeken hoe bepaalde maatregelen deze verspreiding kunnen tegengaan, zijn wiskundige modellen onze enige optie, daar het natuurlijk niet mogelijk is de verschillende mogelijkheden op grote schaal in werkelijkheid uit te proberen. Virtuele experimenten, waarbij we het model oplossen voor parameterwaarden die aangepast werden naargelang de bestudeerde maatregelen, leveren hier levensnoodzakelijke informatie.




##  T6.2 - Soorten modellen


###  T6.2.1 - Algebraïsche vergelijkingen

In algebraïsche vergelijkingen drukt men de waarde van de afhankelijke veranderlijke gewoon uit in functie van de relevante onafhankelijke veranderlijken, parameters en constanten; er komen geen afgeleiden aan te pas. In [H1 - Farmacokinetiek](./H1-Farmacokinetiek.ipynb) beginnen we het hoofdstuk met een algebraïsche vergelijking, 

$$
		c_p = \dfrac{D}{V_d} e^{-k t}.
$$

Later blijkt dat we deze algebraïsche vergelijking bekomen door een differentiaalvergelijking op te lossen.

###  T6.2.2 - Differentiaalvergelijkingen

Differentiaalvergelijkingen drukken *de verandering* van een kwantiteit uit in functie van de kwantiteit zelf, en de relevante onafhankelijke veranderlijken, parameters en constanten. Deze verandering schrijven we onder de vorm van een afgeleide, ook wel *differentiaal* genoemd.

Later in dit hoofdstuk verduidelijken we dat de differentiaalvergelijking zich er uitstekend toe leent om de verandering van een kwantiteit te linken aan chemische, fysische en biologische wetmatigheden. Daarom wordt deze ook uitvoerig gebruikt om modellen op te stellen.


###  T6.2.3 - Partiële differentiaalvergelijkingen

Een partiële differentiaalvergelijking is een differentiaalvergelijking die afhankelijk is van meer dan één onafhankelijke veranderlijke. We komen hiermee in [H7-Diffusie](./H7-Diffusie.ipynb) kort mee in aanraking. Inderdaad, indien we ons voorstellen dat we de diffusie van een druppel inkt in een glas water willen modelleren, lijkt het ons gewenst om voor elke plaats in het glas water de verandering van de concentratie inkt doorheen de tijd neer te schrijven. Ons model zal dus zowel van de plaats als de tijd afhankelijk zijn.

###  T6.2.4 - Statistische modellen

Bovenstaande modellen, en de meeste modellen uit deze cursus, gaan uit van chemische, fysische en biologische principes om de werkelijkheid tot een model te abstraheren, en vervolgens aan de hand van experimentele data na te gaan of het resultaat de realiteit nabootst. Statistische modellen werken op een heel andere manier. Men verzamelt gelijktijdig experimentele data voor de kwantiteit die men wil modelleren, en voor andere kwantiteiten waarvan men denkt dat ze een invloed hebben op de te modelleren kwantiteit. Vervolgens gaat men met statistische technieken correlaties zoeken tussen de data voor de te modelleren kwantiteit en de andere kwantiteiten. 

In [T7-Lineaire Regressie](./T7-LineaireRegressie) beschouwen we een van de meest eenvoudige statistische modellen. Hier hebben we één te modelleren kwantiteit, en één andere kwantiteit. Ons model onderstelt dat er een lineaire relatie bestaat tussen de twee, en men probeert dus een rechte te vinden die de te modelleren kwantiteit aan de andere kwantiteit linkt.

Verder bouwend op deze ideëen kunnen we op basis van de gigantische datasets en de sterke computers die vandaag voorhanden zijn, zeer ingewikkelde statistische modellen bouwen. De zogenaamde *machine learning* en *artificiële intelligentie* zijn veelal op deze principes gebaseerd. Deze technieken stellen ons in staat modellen te bouwen voor problemen waar er weinig hoop is om dit zelf op basis van chemische, fysische en biologische principes te doen.



##  T6.3 - Modelleren

Het opstellen van een model noemt men ook het modelleren. Hoe begint men hier nu aan? We bespreken hier de modus operandi die gehanteerd wordt wanneer men differentiaalvergelijkingen gebruikt, zoals in deze cursus het geval is. Andere modellen zullen volgens dezelfde ideëen opgesteld worden, maar op sommige momenten valt er een onderscheid te maken, wegens technische of filosofische verschillen.

###  T6.3.1 - Wat willen we modelleren?

Het is belangrijk eerst precies vast te leggen welk onderdeel van de realiteit we willen modelleren, en wat we willen doen met ons model. Slechts zo kunnen we assumpties kiezen die de situatie voldoende vereenvoudigen, maar toch ruimte laten voor de precisie die onze vraagstelling verwacht. We bakenen het object van onze modelleerqueeste, ook wel het *systeem* genoemd, zorgvuldig af, zodat we weten waar ons systeem eindigt, en de rest van de realiteit begint. Het systeem wordt vaak diagrammatisch als een afgebakende rechthoek in de realiteit weergegeven.

<img src="figs/model/model_leeg.png" width="150" />

Stel vast dat ook deze stap reeds met het maken van assumpties gepaard gaat. We kiezen namelijk welke factoren binnen ons systeem liggen, en dus welke factoren we aandachtig zullen bestuderen en modelleren. De rest van de realiteit ligt buiten het systeem, en zal (ten dele) verwaarloosd worden.


###  T6.3.2 - Productie en transport

Vaak bestaat onze opdracht eruit de evolutie van een bepaalde kwantiteit doorheen de tijd te modelleren. Vaak is de praktijk dan, en bij uitstek wanneer we differentiaalvergelijkingen wensen te hanteren, om na te gaan hoe deze kwantiteit binnen het systeem veranderd kan worden. Er zijn maar twee opties: ofwel wordt de kwantiteit binnen het systeem geproduceerd of vernietigd, ofwel wordt ze het systeem in of uit getransporteerd.

<img src="figs/model/model_algemeen.png" width=350 />

Men maakt dan een lijst van de manieren waarop de kwantiteit getransporteerd of geproduceerd kan worden. Ook hier komen assumpties bij kijken, omdat bepaalde productie- of transportprocessen mogelijks te verwaarlozen zijn.


###  T6.3.3 - Chemische, fysische en biologische principes

De vraag is nu hoe we deze productie- of transportprocessen wiskundig kunnen uitdrukken. Hier baseren we ons op wetenschappelijke inzichten, die deze processen tot chemische, fysische en biologische wetmatigheden en principes herleid hebben en die, onder bepaalde voorwaarden en assumpties, als eenvoudige wiskundige vergelijkingen kunnen worden gemodelleerd. We construeren op basis hiervan, rekening houdend met de gemaakte assumpties, de relevante vergelijkingen voor alle opgelijste productie- of transportprocessen, en tellen deze bij elkaar op om uit te drukken hoe de kwantiteit verandert doorheen de tijd:

$$
\dfrac{\textit{verandering kwantiteit}}{\textit{tijd}} = \dfrac{\textit{productie}}{\textit{tijd}} + \dfrac{\textit{transport}}{\textit{tijd}}.
$$


###  T6.3.4 - Oplossen van de differentiaalvergelijking

We zijn natuurlijk niet geïnteresseerd in de verandering van de kwaliteit doorheen de tijd, maar in de kwantiteit zelf. Hiertoe lossen we de differentiaalvergelijking op, zodat we een uitdrukking voor de waarde van de kwantiteit op elk tijdstip vinden:

$$
\dfrac{\textit{verandering kwantiteit}}{\textit{tijd}} \rightarrow \textit{kwantiteit}(t).
$$

Hiervoor zijn verschillende methodes voorhanden, maar we beschouwen in deze cursus slechts de eenvoudigste gevallen (zie [T5 - Differentiaalvergelijkingen](./T5-Differentiaalvergelijkingen.ipynb)). Het exact oplossen van differentiaalvergelijkingen is vaak een erg moeilijke, en meestal zelfs onmogelijke taak. Daarom dienen we gebruik te maken van computers, die in staat zijn ons een benaderende versie te geven van de oplossing die geen mens ooit zal kennen (zie bijvoorbeeld [H3 - Populatiemodellen](./H3-Populatiemodellen.ipynb)).

Een differentiaalvergelijking vertelt ons hoe een kwantiteit verandert doorheen de tijd. Om op basis hiervan de waarde van de kwantiteit op verschillende tijdstippen te berekenen, moeten we de waarde op een begintijdstip kennen. Slechts dan kunnen we, op basis van deze waarde en de daaropvolgende verandering, de waarde op het volgend tijdstip berekenen, en zo voort. Deze waarde op een begintijdstip noemen we de *initiële conditie*. Er is steeds een initiële conditie nodig om een differentiaalvergelijking op te kunnen lossen.


###  T6.3.5 - Parameterschatting

In wiskundige modellen komen chemische, fysische of biologische constanten en parameters voor. De waarden hiervan kunnen soms in de literatuur of door middel van andere eenvoudige experimenten gevonden worden, maar soms zijn ze specifiek aan ons model, en moeten we ze *schatten*. Hiertoe voeren we enkele experimenten uit, en gaan we de kwantiteit binnen het systeem dat we bestuderen op verschillende tijdstippen meten, om zodoende experimentele data te verzamelen. Vervolgens gaat men de modelvoorspelling, dus de oplossing van het model, vergelijken met deze experimentele data, om te schatten welke parameterwaarde men in het model zou moeten invullen opdat het inderdaad overeen zou stemmen met de geobserveerde data. Vaak komt het erop neer die parameterwaarde te vinden waarvoor het modelvoorspelling zo dicht mogelijk bij de experimentele data aanleunt. In eenvoudige gevallen kunnen we hiervoor een [lineaire regressie](./T7-LineaireRegressie) gebruiken.

###  T6.3.6 - Vergelijken met de realiteit

Ons model is af, maar het is nu belangrijk te onderzoeken of het ook in staat is een voorspelling van aanvaardbare kwaliteit te doen over de realiteit. We verzamelen daarom nieuwe data, onder verschillende nieuwe experimentele omstandigheden, en vergelijken de werkelijkheid met de modelvoorspelling. Op basis van de kwaliteit van de overeenkomst beslissen we of ons model waardevol is. Indien niet, moeten we onderzoeken of we misschien te strenge assumpties gemaakt hebben, en belangrijke factoren hebben weggelaten. Indien er wel een goede overeenkomst is, kan het waardevol zijn om te onderzoeken of we nog strengere assumpties kunnen maken. Misschien kunnen onderdelen van ons huidig model verwaarloosd worden zonder significant slechtere modelvoorspellingen te produceren, en kan ons model op die manier eenvoudiger worden gemaakt?
